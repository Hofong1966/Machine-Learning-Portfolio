```python
import pandas as pd
```

```python
dataset = pd.read_csv("Wine-quality-challenge.csv")
dataset.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
n_components = np.arange(1,10)
models = [GaussianMixture(n,random_state=1502).fit(dataset) 
         for n in n_components]
plt.plot(n_components,
         [m.bic(dataset) for m in models],
         label='BIC')
plt.plot(n_components,
         [m.aic(dataset) for m in models],
         label='AIC')
plt.legend(loc='best')
plt.xlabel('number of components')
```

```python
#GMM
model = GaussianMixture(n_components = 7,
                        random_state = 1502).fit(dataset)
```

```python
# interpretation
interpretation = pd.DataFrame(model.means_,columns = list(dataset.columns))
interpretation
```
